```python
# Merging ODI Data 

# Merging on Player
odi_data = odibat.merge(odibowl, on='Player',how = "outer",suffixes = ('_bat', '_bowl'))
odi_data = odi_data.merge(odifield, on='Player',how = "outer")

# Adding a "Format" column
odi_data['Format'] = 'ODI'
```

```python
# Merging T20 Data

t20_data = t20bat.merge(t20bowl, on="Player", how="outer", suffixes=('_bat', '_bowl'))
t20_data = t20_data.merge(t20field, on="Player", how="outer")

t20_data["Format"] = "T20"
```

```python
# Merging Test Data

test_data = testbat.merge(testbowl, on="Player", how="outer", suffixes=('_bat', '_bowl'))
test_data = test_data.merge(testfield, on="Player", how="outer")

test_data["Format"] = "Test"
```

```python
final_data = pd.concat([odi_data, t20_data, test_data], ignore_index=True)
```
